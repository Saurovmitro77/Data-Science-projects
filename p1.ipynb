```python
l = []
for i in df['species']:
    if i == 'setosa':
        l.append(1)
    elif i == 'versicolor':
        l.append(2)
    else:
        l.append(3)
```

```python
df['encoded'] = l #feature engineering
```

```python
df.head()
```

```python
df.shape
```

```python
df.tail()
```

```python
df = df.drop(['species'],axis=1)
```

```python
df.head()
```

```python
x = df[['sepal_length','sepal_width','petal_length','petal_width']]
y = df[['encoded']]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
x_train.shape
```

```python
linear = LinearRegression()
```

```python
linear.fit(x_train,y_train)
```

```python
pred = linear.predict(x_test)
```

```python
pred
```

```python
r2_score(y_test,pred)
```

```python
linear.coef_
```

```python
linear.intercept_
```
